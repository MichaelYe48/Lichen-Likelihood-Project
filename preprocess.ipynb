```python
import pandas as pd
from pathlib import Path

# Load the CSV file into a DataFrame
csv_path = Path('air_lichen_query.csv')
df = pd.read_csv(csv_path, engine='python', on_bad_lines='skip')
```

```python
# Get statistics for selected columns and bin the element columns
import numpy as np
cols = ['Magnesium (ppm dw)', 'Potassium (ppm dw)', 'Sodium (ppm dw)', 'Region']
df[cols[:3]].describe()

```

```python
df['Region'].value_counts() # something weird here
```

```python
# BINNING CODE 
for col in cols[:3]:
    data_mean = df[col].mean()
    data_std = df[col].std()
    bins = [float('-inf'), data_mean - (0.25 * data_std), data_mean + (0.25 * data_std), float('inf')]
    if col == "Sodium (ppm dw)":
        bins = [float('-inf'), data_mean - (0.25 * data_std), data_mean + (0.25 * data_std), float('inf')]
    labels = ['low', 'medium', 'high']
    df[f'{col}_binned'] = pd.cut(df[col], bins=bins, labels=labels)
df[[*cols, *(f'{c}_binned' for c in cols[:3])]].head()
```
